# Correlation with python, numpy

### covariance
> 공분산

*X의 편차와 Y의 편차를 곱한 것의 평균*

* 확률변수가 2개 일 때의 선형관계를 나타냄
* 공분산 Cov(x,y) > 0 이면 서로 양의 선형(linear)관계
* 공분산 Cov(x,y) < 0 이면 서로 음의 선형(linear)관계

### correlation
> 상관계수


* 공분산은 두 변수의 scale 값에 따라 값이 달라짐
    * 공분산이 크다고 상관관계가 큰게 아니다.
* 두 변수의 상관관계를 객관화하여 나타내는 계수이다.

* 상관계수에 따른 두 확률변수의 분포

* Exactly **–1.** A perfect downhill (negative) linear relationship<br>

* **-0.70.** A strong downhill (negative) linear relationship
* **-0.50.** A moderate downhill (negative) relationship
* **-0.30.** A weak downhill (negative) linear relationship

<br>

* **0.** No linear relationship

<br>

* **+0.30.** A weak uphill (positive) linear relationship
* **+0.50.** A moderate uphill (positive) relationship
* **+0.70.** A strong uphill (positive) linear relationship<br>

* Exactly **+1.** A perfect uphill (positive) linear relationship

```python
def de_mean(x):
    """translate x by subtracting its mean (so the result has mean 0)"""
    x_bar = np.mean(x)
    return [x_i - x_bar for x_i in x]

def covariance(x,y):
    n = len(x)
    return np.dot(de_mean(x), de_mean(y)) / (n-1)

def correlation(x,y):
    stdev_x = np.std(x)
    stdev_y = np.std(y)
    if stdev_x > 0 and stdev_y > 0:
         return covariance(x, y) / stdev_x / stdev_y
    else:
         return 0 
```

```python
#공분산
print(covariance(num_friends, daily_minutes)) # 22.43
#상관계수
print(correlation(num_friends, daily_minutes)) # 0.25
```

# Covariance and Correlation with numpy

```python
num_friends = np.array(num_friends)
daily_minutes = np.array(daily_minutes)
data = np.array([num_friends, daily_minutes])
print(data.shape)
```

### 공분산 행렬

```python
print(np.cov(data))
```

### 공분산 반환
* 공분산 행렬이기 때문에 첫번째 행의 두번째 열 값이 공분산

```python
print(np.cov(data)[0,1])
```

### 상관계수 행렬

```python
print(np.corrcoef(data))
```

### 상관계수 반환
* 상관계수 행렬이기 때문에 첫번째 행의 두번째 열 값이 상관계수

```python
print(np.corrcoef(data))
```

```python
print(np.corrcoef(data)[0,1])
```

## Example

절대값을 취해줬을 때 상관관계가 없어지지만, 절대값을 취하지 않으면 상관관계가 있다.

```python
# absolute value relationship
x = [-2, -1, 0, 1, 2]
y = [ 2, 1, 0, 1, 2]
plt.scatter(x, y)
plt.show()
```

```python
x = [-2, -1, 0, 1, 2]
y = [99.98, 99.99, 100, 100.01, 100.02]
```

> y축의 변화가 거의 없지만, 변화가 있어보인다.
(y 축에 log 값이 취해짐)

```python
plt.scatter(x, y)
plt.title('corr={:.2}'.format(correlation(x,y)))
plt.show()
```

y 축을 실제값으로 설정했더니, y 값의 변화가 거의 없는 것이 나타나기 때문에 상관관계가 높더라도 흥미로운 상관관계는 아니다.

```python
#Perfectly correlated. This relationship isn’t all that interesting.
x = [-2, -1, 0, 1, 2]
y = [99.98, 99.99, 100, 100.01, 100.02]
plt.scatter(x, y)
plt.axis('equal')
plt.title('corr={:.2}'.format(correlation(x,y)))
plt.show()
```
